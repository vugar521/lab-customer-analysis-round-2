```python
data = data.drop('Vehicle_Type', axis=1) #dropping Vehicle_Type
data.head()
```

```python
data.columns
```

```python
# I have dropped "Vehicle_Type" column at all as it had 5482 NAN values and filled with Median for 'Months_Since_Last_Claim' and 'Number_of_Open_Complaints'.
#Filled the cell containing NaN values with previous entry for rest of columns
```

```python
data['State'] = data['State'].fillna(method='ffill').fillna(method='bfill')#Filling the cell containing NaN values with previous entry
data['Response'] = data['Response'].fillna(method='ffill').fillna(method='bfill')
data['Vehicle_Class'] = data['Vehicle_Class'].fillna(method='ffill').fillna(method='bfill')
data['Vehicle_Size'] = data['Vehicle_Size'].fillna(method='ffill').fillna(method='bfill')
data['Months_Since_Last_Claim'] = data['Months_Since_Last_Claim'].fillna(data['Months_Since_Last_Claim'].median())
data['Number_of_Open_Complaints'] = data['Number_of_Open_Complaints'].fillna(data['Number_of_Open_Complaints'].median())
data.isnull().sum()
```
